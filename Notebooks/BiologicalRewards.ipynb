```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = ""

import gym
import numpy as np
import tensorflow as tf
from narla.man import MultiAgentNetwork
from narla.monitor import Monitor
import matplotlib.pyplot as plt 
```

```python
from IPython.core.pylabtools import figsize
figsize(20, 5)
```

```python
env = gym.make("CartPole-v1")
state = tf.convert_to_tensor(
    env.reset().reshape(1, -1),
    dtype=tf.float32
)
man = MultiAgentNetwork(
    input_size=4, 
    num_layers=4, 
    num_nodes_per_layer=15
)
man(state)
man(state)
man(state)
```

```python
for i, con in enumerate(man.layer_connectivity):
    plt.subplot(1, 5, i+1)
    plt.imshow(con, vmin=0, vmax=1, aspect='equal')
    plt.grid(which='major', axis='both', linestyle='-')
    plt.tight_layout()
    plt.title(str(con.shape))
```

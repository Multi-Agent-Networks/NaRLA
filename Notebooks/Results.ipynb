```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
from IPython.core.pylabtools import figsize
figsize(20, 8)
```

```python
df = pd.read_csv('../results.csv')
```

```python
for i, num_nodes in enumerate(df.num_nodes.unique()):
    plt.subplot(2,2,i+1)
    df_nodes = df[df.num_nodes == num_nodes]
    sns.lineplot(x='Episode', y='Rewards', data=df_nodes, hue='num_layers')
    plt.title(f'{num_nodes} Nodes'); plt.tight_layout()
```

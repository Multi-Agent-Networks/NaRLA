# Single Node Training

```python
import sys
sys.path.append('..')

import tensorflow as tf
from narla.node import Node

tf.random.set_seed(123)
```

```python
node = Node()
```

```python
# Training data
X = tf.random.uniform((100, 3))
Y = tf.random.uniform((100, 1))
```

### Storing seperate tapes

```python
for i in range(10):
    tapes = []
    outputs = []

    for x in X:
        _x = tf.reshape(x, (1, 3))
        with tf.GradientTape(persistent=True) as tape:
            outputs.append(node(_x)[0])
            tapes.append(tape)

    opt = tf.keras.optimizers.SGD(learning_rate=0.01)
    losses = 0

    for tape, output, y in zip(tapes, outputs, Y):
        tape.__enter__()
        loss = (output - y)**2
        tape.__exit__(None, None, None)
        
        losses += loss
        
    grads = tape.gradient(loss, node.trainable_variables)
    opt.apply_gradients(
        zip(grads, node.variables)
    )

    print(
        losses
    )
```

### One Tape : Calculate loss immediately

```python
opt = tf.keras.optimizers.SGD(learning_rate=0.01)
for i in range(10):
    losses = 0
    for x, y in zip(X, Y):
        _x = tf.reshape(x, (1, 3))
        with tf.GradientTape(persistent=True) as tape:
            output = node(_x)[0]        
            loss = (output - y)**2

        losses += loss
        
    grads = tape.gradient(loss, node.trainable_variables)
    opt.apply_gradients(
        zip(grads, node.variables)
    )

    print(
        losses
    )
```
